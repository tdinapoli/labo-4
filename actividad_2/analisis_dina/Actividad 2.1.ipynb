```python
import numpy as np
import matplotlib.pyplot as plt

E = 10e10 #N/m
L = 0.4 #m
d = 0.005 #m
m = 0.01 #kg
I = np.pi * d**4 / 64 #m
F = 9.8 * m #N
```

```python
def y(x, E,I,F,L):
    num = F * (L * x**2 - x**3 / 3)
    den = 2*E*I
    return num/den
print("El desplazamiento de la barra para x=L es ~", y(L, E, I, F, L), "m")
print("Lo que en micrmetros es ~", y(L, E, I, F, L)/10e-6, "m")
```

```python
x = np.arange(0,0.5,0.01)
plt.plot(x, -y(x, E,I,F,L))
plt.xlabel("Longitud en x [m]")
plt.ylabel("Desplazamiento en y [m]")
plt.title("Desplazamiento de la barra por un peso de 0.01kg")
plt.show()
```
