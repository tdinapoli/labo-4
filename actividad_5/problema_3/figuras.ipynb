### 1) PSD

```python
import numpy as np
import matplotlib.pyplot as plt

R_std = np.loadtxt('datos_2805/Rstd_p3_fc19_s11_o0.txt')
R_grande_std = np.loadtxt('datos_2805/Rstd_grande_p3_fc19_s11_o0.txt')
frecs = np.logspace(np.log10(10), np.log10(50000), 200, dtype=int)
frecs_grande = np.logspace(np.log10(40000), np.log10(100000), 50, dtype=int)

fig, ax = plt.subplots(1, figsize=(15,8))

ax.plot(frecs, R_std, "ok")
ax.plot(frecs_grande, R_grande_std, "ok")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frecuencia [Hz]", fontsize=20)
ax.set_ylabel("$\sigma_R$ [V]", fontsize=20)
ax.set_title("Ruido en función de la frecuencia", fontsize=25)
ax.tick_params(labelsize=15)
ax.grid()
plt.show()
```

```python
#importo datos de medición sin señal
R_std_sin_ref = np.loadtxt('datos_2805/Rstd_p3_fc19_s12_o0_sin_ref.txt')
frecs_sin_ref = np.logspace(np.log10(10), np.log10(50000), 50, dtype=int)

#defino espacios donde voy a plotear la aproximación lineal
m = -1.0054693351135677
b = -5.922594723336171
x = np.linspace(10, 1e3, 2)
A = m
B = 10**b
y = B * (x**A)


fig, ax = plt.subplots(1, figsize=(15,8))

#ploteo el ruido con señal
ax.plot(frecs, R_std, "ok", label="Alimentado")
ax.plot(frecs_grande, R_grande_std, "ok")

#ploteo ruido sin señal
ax.plot(frecs_sin_ref, R_std_sin_ref, "o", color="#103f79",label="Sin alimentar")

#ploteo aproximación lineal
ax.plot(x, y, color="#f3b229", linewidth=2.5, label="Aproximación lineal")

#parámetros y labels
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Frecuencia [Hz]", fontsize=20)
ax.set_ylabel("$\sigma_R$ [V]", fontsize=20)
ax.set_title("Ruido en función de la frecuencia", fontsize=25)
ax.tick_params(labelsize=15)
ax.grid()
ax.legend(fontsize=15)
plt.show()
```
