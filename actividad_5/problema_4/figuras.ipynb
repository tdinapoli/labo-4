### Linealidad

```python
import numpy as np
import matplotlib.pyplot as plt

R_m_sN = np.loadtxt('datos_2805/Rmean_p4_fc8_s17_o3_tsleep15_sN.txt')
amplitudes = np.arange(0.05, 4.95, 0.05)

fig, ax = plt.subplots(1, figsize=(15,8))

ax.plot(amplitudes, R_m_sN, 'ok')
ax.grid()
ax.set_xlabel("Amplitud entregada [V]", fontsize=15)
ax.set_ylabel("Amplitud recibida [V]", fontsize=15)
ax.tick_params(labelsize=15)
plt.show()
```

### Se√±al/Ruido

```python
R_m_sN = np.loadtxt('datos_2805/Rmean_p4_fc8_s17_o3_tsleep15_sN.txt')
R_std_sN = np.loadtxt('datos_2805/Rstd_p4_fc8_s17_o3_tsleep15_sN.txt')
R_fino_m_sN = np.loadtxt('datos_2805/Rmean_fino_p4_fc8_s17_o3_tsleep15_sN.txt')
R_fino_std_sN = np.loadtxt('datos_2805/Rstd_fino_p4_fc8_s17_o3_tsleep15_sN.txt')

amp = np.arange(0.05, 4.95, 0.05)
amp_fino = np.arange(0.005, 0.2, 0.002)

fig, ax = plt.subplots(1, figsize=(15,8))

ax.plot(amp, R_m_sN/R_std_sN, 'ok')
ax.plot(amp_fino, R_fino_m_sN/R_fino_std_sN, 'ok')
ax.set_yscale("log")
ax.set_ylim([1e1,1e5])
ax.set_title("Se√±al/Ruido", fontsize=25)
ax.set_xlabel("Amplitud de entrada [V]", fontsize=20)
ax.set_ylabel("Amplitud de salida/Desviaci√≥n estandar", fontsize=20)
ax.tick_params(labelsize=20)
ax.grid()
plt.show()
```

```python
%matplotlib inline
T_m_sN_ = np.loadtxt('datos_2805/Tmean_p4_fc8_s17_o3_tsleep15_sN.txt')
T_m_fino_sN_ = np.loadtxt('datos_2805/Tmean_fino_p4_fc8_s17_o3_tsleep15_sN.txt')

T_m_sN = []
T_m_fino_sN =[]
for i in range(len(T_m_sN_)):
    if (i % 2) == 0:
        T_m_sN.append(T_m_sN_[i])
        T_m_fino_sN.append(T_m_fino_sN_[i])

fig, ax = plt.subplots(1, figsize=(15,8))

ax.plot(amp, np.array(T_m_sN) * 2*np.pi /360, 'ok')
ax.plot(amp_fino, np.array(T_m_fino_sN) * 2*np.pi /360, 'ok')
ax.set_xlabel("Amplitud de entrada [V]", fontsize=20)
ax.set_ylabel("Fase de salida [Rad]", fontsize=20)
ax.set_title("Fase en funci√≥n de amplitud de entrada")
ax.tick_params(labelsize=15)
ax.grid()
plt.show()
```
