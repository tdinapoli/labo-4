### Determinacin del orden del filtro

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib widget

R_0 = np.loadtxt('datos_viejos/actividad_2/barrido_R_s12_f8_o0.txt') #Radio, sensibilidad 12, tint  = 8, orden del filtro = 0
R_1 = np.loadtxt('datos_viejos/actividad_2/barrido_R_s12_f8_o1.txt') #Radio, sensibilidad 12, tint  = 8, orden del filtro = 1
R_2 = np.loadtxt('datos_viejos/actividad_2/barrido_R_s12_f8_o2.txt') #Radio, sensibilidad 12, tint  = 8, orden del filtro = 2
R_3 = np.loadtxt('datos_viejos/actividad_2/barrido_R_s12_f8_o3.txt') #Radio, sensibilidad 12, tint  = 8, orden del filtro = 3

plt.plot(R_0, 'b.')
plt.plot(R_1, 'r.')
plt.plot(R_2, 'g.')
plt.plot(R_3, 'k.')
plt.show()
```

En todos los casos se obserba que hay un pico negativo en el voltaje de salida, esto no es lo esperado, pero salvo en ese punto que est al rededor de 100 Hz se comporta como esperamos para los filtros de orden 1,2 y 3 no es as para el filtro de orden 0.

## Diagramas de Bode

```python
frecs = np.logspace(np.log10(40), np.log10(50000), dtype=int)

R = np.loadtxt('datos_viejos/actividad_2/barrido_R_s12_f10_o3.txt')
tita = np.loadtxt('datos_viejos/actividad_2/barrido_titas_s12_f10_o3.txt')
Transf = R*2/0.1

plt.figure()
plt.plot(frecs, tita, '.')
plt.ylabel('fase $\phi$')
plt.xlabel('frecuencias [Hz]')
plt.xscale('log')
plt.grid()

plt.figure()
plt.plot(frecs, Transf, '.')
plt.xlabel('frecuencias [Hz]')
plt.ylabel('Voltaje de salida/2 [V]')
plt.xscale('log')
plt.grid()
plt.show()
```

```python
%matplotlib widget

R      = np.loadtxt('datos/R_p2_fc8_s12_o3.txt')
R_fino = np.loadtxt('datos/R_p2__fino_fc8_s12_o3.txt')
T      = np.loadtxt('datos/T_p2_fc8_s12_o3.txt')
T_fino = np.loadtxt('datos/T_p2_fino_fc8_s12_o3.txt')

frecs      = np.logspace(np.log10(40), np.log10(50000), 100, dtype=int)
frecs_fino = np.arange(82, 131, 1, dtype=int)

Transferencia = 2*R/0.1
Transferencia_fino = 2*R_fino/0.1


fig, ax  = plt.subplots(1,2, figsize =(10,5),facecolor='white')

ax[0].plot(frecs,  np.log10(Transferencia), '.b', label = 'Barrido entre 40 y 50000')
ax[0].plot(frecs_fino, np.log10(Transferencia_fino), '.r', label = 'Barrido entre 82 y 130')
ax[0].set_xscale('log')
ax[0].set_xlabel('frecuencias [Hz]')
ax[0].set_ylabel('Transferencia [db]')
ax[0].legend()

ax[1].plot(frecs, T, 'b.')
ax[1].plot(frecs_fino, T_fino, 'r.')
ax[1].set_xscale('log')
ax[1].set_xlabel('frecuencias [Hz]')
ax[1].set_ylabel('fase $\phi$ [grad]')
```
